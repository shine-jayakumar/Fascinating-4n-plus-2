# 4n+2

As if the function, 4n+2, wasn't fascinating enough already, I recently stumbled upon a random multiple choice question

The question was:

Which one of the following is a multiple of 4n+2, for n = 1,2,3,4,...

a) 2

b) 3

c) 5

d) 6

Now, for 4n+2 to be a multiple of a number x, if you divide (4n+2) by x, the remainder should be 0

It was simple already, but I decided to write a function that tells me if a number (4n+2, in our case) is a multiple of another number (x).


```python
# check if a given number is a multiple of a number
def is_multiple(num, mul_of):
    return num%mul_of == 0
```

<br/>

### Let's try option a) 2

```python
limit = 25
mul_of = 2

for n in range(1, limit+1):
    print(f"For n={n}, 4n+2={4*n+2}. Is 4n+2 a multiple of {mul_of}: {is_multiple(4*n+2, mul_of)}")
    
```

### So, we now know that 2 satisfies the condition.

### How about 3?

```python
limit = 25
mul_of = 3

for n in range(1, limit+1):
    print(f"For n={n}, 4n+2={4*n+2}. Is 4n+2 a multiple of {mul_of}: {is_multiple(4*n+2, mul_of)}")
```

### How about 5?

```python
limit = 25
mul_of = 5

for n in range(1, limit+1):
    print(f"For n={n}, 4n+2={4*n+2}. Is 4n+2 a multiple of {mul_of}: {is_multiple(4*n+2, mul_of)}")
```

Of course, we know that it doesn't satisfy the condition but I urge you to look closely.

**There's a pattern in which True and False appear**. Also, it's different for different numbers.

Let's create a function that gives us x and y required for plotting where,

x is 1,2,3,4...limit

y is 1 (True) and 0 (False) based on whether 4n+2 is a multiple of given number or not.

```python
def coord_4n_plus_2(num, limit):
    x = [n for n in range(1, limit+1)]
    y = [int(is_multiple(4*n+2, num)) for n in x]
    return [x,y]
    
```

```python
import matplotlib.pyplot as plt
```

### Plotting  with 2

```python
x,y = coord_4n_plus_2(2, 50)

plt.plot(x,y)
plt.show()
```

### Plotting  with 3

```python
x,y = coord_4n_plus_2(3, 100)

plt.figure(figsize=(10, 2))
plt.plot(x,y)
plt.show()
```

### Plotting  with 9

```python
x,y = coord_4n_plus_2(9, 100)

plt.figure(figsize=(10, 2))
plt.plot(x,y)
plt.show()
```

## Plotting primes

### Prime number - 29

```python
x,y = coord_4n_plus_2(29, 10000)

plt.figure(figsize=(20, 2))
plt.plot(x,y)
plt.title("Is 4n+2 a multiple of 29?")
plt.show()
```

Do you know what amazes me? That pattern looks very similar to the interference pattern in the [double-slit experiement](https://en.wikipedia.org/wiki/Double-slit_experiment)

Let's reduce the figure a bit

```python
x,y = coord_4n_plus_2(29, 10000)

plt.figure(figsize=(15, 2))
plt.plot(x,y)
plt.title("Is 4n+2 a multiple of 29?")
plt.show()
```
